```python
import tkinter as tk
from tkinter import messagebox
import random
from queue import Queue

```

```python
class SearchBotApp:
    def __init__(self, master):
        self.master = master
        self.master.title("5x5 Grid Search Bot")
        
        # Initialize grid
        self.grid_size = 5
        self.grid = [[random.randint(1, 100) for _ in range(self.grid_size)] for _ in range(self.grid_size)]
        
        # Create UI
        self.buttons = [[None for _ in range(self.grid_size)] for _ in range(self.grid_size)]
        for i in range(self.grid_size):
            for j in range(self.grid_size):
                self.buttons[i][j] = tk.Button(master, text=str(self.grid[i][j]), width=10, height=3)
                self.buttons[i][j].grid(row=i, column=j)
        
        # Search value input
        self.search_label = tk.Label(master, text="Enter value to search (1-100):")
        self.search_label.grid(row=self.grid_size, column=0, columnspan=2)
        self.search_entry = tk.Entry(master)
        self.search_entry.grid(row=self.grid_size, column=2, columnspan=2)
        self.search_button = tk.Button(master, text="Search", command=self.start_search)
        self.search_button.grid(row=self.grid_size, column=4, columnspan=1)
    
    def start_search(self):
        try:
            target_value = int(self.search_entry.get())
            if 1 <= target_value <= 100:
                self.bfs_search(target_value)
            else:
                messagebox.showerror("Error", "Please enter a number between 1 and 100.")
        except ValueError:
            messagebox.showerror("Error", "Please enter a valid number.")

    def bfs_search(self, target_value):
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        direction_symbols = ["→", "↓", "←", "↑"]
        start_pos = (0, 0)
        queue = Queue()
        queue.put(start_pos)
        visited = set()
        visited.add(start_pos)
        
        while not queue.empty():
            current_pos = queue.get()
            x, y = current_pos
            self.update_button_color(x, y, "yellow")
            self.master.update()
            self.master.after(500)  # Delay for visualization
            
            if self.grid[x][y] == target_value:
                self.update_button_color(x, y, "green")
                messagebox.showinfo("Success", f"Found {target_value} at position ({x}, {y})")
                return
            
            for idx, direction in enumerate(directions):
                new_x, new_y = x + direction[0], y + direction[1]
                if 0 <= new_x < self.grid_size and 0 <= new_y < self.grid_size and (new_x, new_y) not in visited:
                    self.update_button_text(x, y, direction_symbols[idx])
                    queue.put((new_x, new_y))
                    visited.add((new_x, new_y))
        
        messagebox.showinfo("Not Found", f"{target_value} not found in the grid.")
    
    def update_button_color(self, x, y, color):
        self.buttons[x][y].config(bg=color)
        
    def update_button_text(self, x, y, text):
        current_text = self.grid[x][y]
        self.buttons[x][y].config(text=f"{current_text}\n{text}")
```

```python

root = tk.Tk()
app = SearchBotApp(root)
root.mainloop()
```
